```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
car=pd.read_csv("Cars93.CSV")
car.head()
```

```python
columns = ["Manufacturer","Model","Type","Price",
"MPG.city","MPG.highway","Horsepower","Rear.seat.room","Passengers"]
```

```python
car[columns].head()
```

```python
#First sort the data to get a proper line chart
car=car.sort_values(by="Horsepower")
#The following lines of code create a blank canvas to plot on
fig, ax = plt.subplots()
fig.set_figwidth(10)
fig.set_figheight(7)
#Data is fed and plotted using the following lines
car.plot(ax = ax, x = "Horsepower", y = "MPG.highway", kind = "line", )
car.plot(ax = ax, x = "Horsepower",y= "MPG.city", kind = "line", linestyle='--')
```

```python
#First sort the data to get a proper line chart
car=car.sort_values(by="Horsepower")
#Plotting
fig = plt.figure()
fig.set_figwidth(10)
fig.set_figheight(7)
plt.plot(car["Horsepower"], car["MPG.city"],label="MPG.city")
plt.plot(car["Horsepower"], car["EngineSize"], label="Engine Size")
plt.plot(car["Horsepower"], car["MPG.highway"], label="MPG.highway")
plt.plot(car["Horsepower"], [i/100 for i in car["RPM"]],label="RPM")
plt.suptitle("Horsepower vs Mileage in city, Engine size, Mileage on highway, RPM",fontsize=16)
plt.xlabel("Horsepower")
plt.ylabel("Mileage in city, Engine size, Mileage on highway, RPM")
plt.legend()
```

```python
fig = plt.figure()
fig.set_figwidth(10)
fig.set_figheight(9)
plt.plot(car["MPG.city"],car["Horsepower"],label="MPG.city")
plt.plot(car["EngineSize"],car["Horsepower"],label="Engine Size")
plt.plot(car["MPG.highway"],car["Horsepower"],label="MPG.highway")
plt.plot([i/100 for i in car["RPM"]],car["Horsepower"],label="RPM") #interchanging x and y
plt.suptitle("Mileage in city, Engine size, Mileage on highway, RPM vs Horsepower",fontsize=15)
plt.ylabel("Horsepower")
plt.xlabel("Mileage in city, Engine size, Mileage on highway, RPM")
plt.legend()
```
