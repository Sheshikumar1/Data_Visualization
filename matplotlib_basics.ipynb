```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.rcParams['figure.figsize']=(4,3)
plt.rcParams['figure.dpi']=150
```

```python
#basic line plot
val=np.arange(0,10,0.1)
plt.plot(val,np.sin(val))
```

```python
plt.plot(val,np.sin(val))
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
#multiple plot
plt.plot(val,np.sin(val))
plt.plot(val,np.cos(val))
```

```python
#Legend and Title
plt.plot(val,np.sin(val),label=r"$y=\sin(x)$")
plt.plot(val,np.cos(val),label=r"$y=\sin(y)$")
plt.legend()
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
```

```python
#colors to line
plt.plot(val,np.sin(val),label=r"$y=\sin(x)$",color="red")
plt.plot(val,np.cos(val),label=r"$y=\sin(y)$",color="green")
plt.legend()
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
```

```python
plt.plot(val,np.sin(val),label=r"$y=\sin(x)$",color="r",linestyle="--")
plt.plot(val,np.cos(val),label=r"$y=\sin(y)$",marker='o',linestyle="none")
plt.legend()
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
```

```python
#short Format code
plt.plot(val,np.sin(val),"r+--",label=r"$y=\sin(x)$")
plt.legend()
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
```

```python
plt.plot(val,np.sin(val),"r+--",label=r"$y=\sin(x)$")
plt.plot(val,np.cos(val),"bo",label=r"$y=\sin(y)$")
plt.xlabel("$x$")
plt.ylim(0,1)
plt.ylabel(r"$y$")
plt.legend()
```

```python
plt.plot(val,np.sin(val),"r+--",label=r"$y=\sin(x)$")
plt.plot(val,np.cos(val),"bo",label=r"$y=\sin(y)$")
plt.xlabel("$x$")
plt.ylim(0,1)
plt.xlim(0,1)
plt.ylabel(r"$y$")
plt.legend()
```

```python
#Generate 1000 random numbers
rand=np.random.normal(size=1000)
```

```python
#histrogram
plt.hist(rand)
plt.xlabel(r'$x$')
plt.ylabel(r'$Pr(x)$')
```

```python
#Generate a smoothed,gaussian random field
from scipy.ndimage import gaussian_filter
rands2d=gaussian_filter(np.random.normal(size=(512,512)),sigma=10)
print(rands2d.shape)
```

```python
#show image
plt.imshow(rands2d,cmap='viridis')
plt.colorbar()
```

```python
plt.imshow(rands2d,cmap='magma')
plt.colorbar()
```

```python
plt.imshow(rands2d,cmap='inferno')
plt.colorbar()
```

```python
#different line styles
idx=0
for marker in ('-','--',':','-.'):
    plt.plot(val, np.sin(val) + idx, 'k' + marker)
    idx+=1
```

```python
#marker frequency(markevery)
val2=np.arange(0,10,0.01)
plt.plot(val2,np.sin(val2),'k.')
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
val2=np.arange(0,10,0.01)
plt.plot(val2,np.sin(val2),'k.',markevery=10)
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
val2=np.arange(0,10,0.01)
plt.plot(val2,np.sin(val2),'r.',markevery=10)
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
val2=np.arange(0,10,0.01)
plt.plot(val2,np.sin(val2),'r.',markevery=10,markersize=20)
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
#different marker
idx=0
for marker in ('.',',','o','v','^','>','<','1','2','3','4','s','p','*','h','H','+','x','D','d','|','_'):
    plt.plot(idx,0,''.join(('r',marker)))
    idx+=1
plt.xlim(-0.5,idx+0.5)
```

```python
plt.plot(val,np.sin(val),color='k',linestyle='none',marker=r'$\Sigma$')
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
plt.plot(val,np.sin(val),color='k',linestyle='none',marker=r'$\sin$')
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
plt.plot(val,np.sin(val),lw=8)
plt.plot(val+1,np.sin(val),lw=8)
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
plt.plot(val,np.sin(val),lw=8,zorder=10)
plt.plot(val+1,np.sin(val),lw=8,zorder=2)
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
#transparency
plt.plot(val,np.sin(val),lw=8,alpha=0.5)
plt.plot(val+1,np.sin(val),lw=8,alpha=0.5)
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
#array order matters
shuffled_val=np.random.permutation(val)
plt.plot(shuffled_val,np.sin(shuffled_val),'k.')
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
shuffled_val=np.random.permutation(val)
plt.plot(shuffled_val,np.sin(shuffled_val),'k--')
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

# Scatter()

```python
vals=np.arange(0,10,0.1)
plt.scatter(vals,np.sin(vals))
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
#colors
plt.scatter(vals,np.sin(vals),color='r')
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
plt.scatter(vals,np.sin(vals),c=np.cos(vals))
plt.colorbar()
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
#colormaps
plt.scatter(vals,np.sin(vals),c=np.cos(vals),cmap='inferno')
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
#marker edge colors
plt.scatter(vals,np.sin(vals),c=np.cos(vals),edgecolor="none")
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
plt.scatter(vals,np.sin(vals),s=1,c=np.cos(vals),edgecolor='k',marker=',',)
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
plt.scatter(vals,np.sin(vals),s=1,c=np.cos(vals),edgecolor='none',marker=',',)
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```
