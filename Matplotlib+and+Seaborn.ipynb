```python
# Importing libraries
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# Creating dummy data
data = pd.DataFrame({
    'A': np.random.randn(100),
    'B': np.random.rand(100),
    'C': np.random.choice(['Group1', 'Group2', 'Group3'], 100)
})
```

```python
data.head(10)
```

```python
data['D'] = 2*data['A'] + 0.05
```

```python
data.head(10)
```

```python
# Common plot types
# Distribution plot
sns.histplot(data['A'],kde=True)
plt.title('Distribution Plot')
plt.show()
```

```python
# Scatter plot
plt.scatter(data['A'], data['B'])
plt.title('Scatter Plot')
plt.xlabel('A')
plt.ylabel('B')
plt.show()
```

```python
# Box plot
sns.boxplot(x='C', y='A', data=data)
plt.title('Box Plot')
plt.show()
```

```python
# Setting figure size, titles, axis names, and limits
plt.figure(figsize=(10, 6))
plt.plot(data['A'])
plt.title('Line Plot with Customizations')
plt.xlabel('Index')
plt.ylabel('Values')
plt.xlim(0, 100)
plt.ylim(data['A'].min(), data['A'].max())
plt.show()
```

```python
# Multiple charts on the same chart with hue
sns.scatterplot(x='A', y='B', hue='C', data=data)
plt.title('Scatter Plot with Hue')
plt.show()
```

```python
# Grouped charts
# Distribution of 2 continuous columns with respect to one categorical variable
sns.histplot(data[data['C'] == 'Group1']['A'], label='Group1',kde = True)
sns.histplot(data[data['C'] == 'Group2']['A'], label='Group2', kde = True)
sns.histplot(data[data['C'] == 'Group3']['A'], label='Group3', kde = True)
plt.legend()
plt.title('Grouped Distribution Plots')
plt.show()
```
